## What is artificial intelligence, machine learning and deep learning?

* Artificial Intelligence (AI) - Any techniques that allow us to mimic human behaviour

* Machine Learning (ML) - Subset of AI and allow the computers to learn from data using statistical methods without being explicitly programmed

* Deep Learning (DL) - Subset of ML which make the computation using multi-layer neural network is feasible

## Why now?

In early days, there are some limitations in term of compute or processing power and limited statistical technique available to create an efficient AI model  
Now the rise of AI has been greatly contributed by the following factors:
+ Big Data - massive data generated by the growth of internet
+ Moore's Law - compute power is increasing and now computer can process large amount of data faster
+ Rapid increase in applications - smartphone, computers and IoT drives the technology

![Moore's_Law](Images/Moore's_Law.png)

Moore's Law: As the scale of chip components gets closer and closer to that of individual atoms, it's gotten harder to keep up the pace of Moore's Law. It's now more expensive and more technically difficult to double the number of transistors -- and thus the processing power -- for a given chip every two years. 



## Types of Machine Learning

Generally there are 4 types of machine learning (some publications mentioned only 3 types):
+ Supervised Learning - Relationship between input and output by using labeled data
+ Unsupervised Learning - Relationship between input and output by using unlabeled data
+ Reinforcement Learning - Agents that take action to maximised rewards
+ Semi-supervised Learning - Relationship between input and output by using mixed of labeled and unlabeled data

## What is deep learning?

Deep learning mimics how our brain works by using deep layered of neural network\
![Neural Network](Images/neural_network.png)

The layers are made of nodes. A node is just a place where computation happens, similar to a neuron in our brain, which fires when it encounters sufficient stimuli. A node combines input from the data with a set of coefficients, or weights, that either amplify or dampen that input, thereby assigning significance to inputs with regard to the task the algorithm is trying to learn. These input-weight products are summed and then the sum is passed through a node’s so-called activation function, to determine whether and to what extent that signal should progress further through the network to affect the ultimate outcome, say, an act of classification. If the signals passes through, the neuron has been “activated.”
